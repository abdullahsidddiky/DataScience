```python
import librosa
import IPython.display as ipd
import matplotlib.pyplot as plt
import librosa.display
import sklearn
from sklearn.preprocessing import minmax_scale
import pandas as pd 
import os
```

```python
%matplotlib inline
```

```python
plt.style.use('fivethirtyeight')
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
audio_files=['a1.wav','a2.wav','a3.wav','a4.wav','a5.wav',
            'b1.wav','b2.wav','b3.wav','b4.wav','b5.wav',
            'c1.wav','c2.wav','c3.wav','c4.wav','c5.wav']
audio_filepaths = ['./data/sample/'+x for x in audio_files]
```

```python
audio_filepaths[6]
```

```python
x,sr = librosa.load(audio_filepaths[6],sr = 44100)
```

```python
len(x)
```

```python
plt.figure(figsize=(14,5))
librosa.display.waveplot(x,sr=sr)
```

```python
x = x
sr = sr
spectral_centroids = librosa.feature.spectral_centroid(x,sr=sr)[0]
print('spectral_centroids:',spectral_centroids.shape)
plt.figure(figsize=(12,4))
frames = range(len(spectral_centroids))
t=librosa.frames_to_time(frames)

def normalize(x,axis= 0):
    return sklearn.preprocessing.minmax_scale(x,axis=axis)
nx = normalize(spectral_centroids)
print(len(nx))
print(nx)
librosa.display.waveplot(x,sr=sr, alpha=0.4)
plt.plot(t,normalize(spectral_centroids),color ='b')
```

```python
x = x
sr = sr
spectral_bandwidth_2=librosa.feature.spectral_bandwidth(x+0.01,sr=sr,p=2)[0]
spectral_bandwidth_3=librosa.feature.spectral_bandwidth(x+0.01,sr=sr,p=3)[0]
spectral_bandwidth_4=librosa.feature.spectral_bandwidth(x+0.01,sr=sr,p=4)[0]

srx = normalize(spectral_bandwidth_2)
print(len(srx))
print(srx)
```
